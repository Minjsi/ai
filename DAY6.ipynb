```python
# 합계값을 계산하는 알고리즘

첨자 index 0  1 2   3   4 
데이터    12 13 11 14  10

데이터의 합계를 구하는 가장 단순한 알고리즘 - sum

arr[0]+arr[1]+arr[2]+arr[3]+arr[4]
로도 표현할 수 있다.
그러나 배열의 개수가 늘어나면 사용하기 힘들다
반복구조로 변형하자

arr[0]->sum
sum+arr[1]->sum
sum+arr[2]->sum
sum+arr[3]->sum
sum+arr[4]->sum
```

```python
arr=[12,13,11,14,10]
sum=0
for i in arr :
        sum=sum+i

print(sum)
```

```python
# Linear Search 선형 탐색법

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다. 
# 탐색 처리는 반복 구조를 사용하여 기술한다. 
# 반복 구조에서는 종료 조건을 잊지 말자. 

첨자 (index)   0  1  2  3  4
데이터          4  2  3  5  1    찾을 데이터  :  5
 
```

```python
#for i in range(0,5) -> index 추출
#for i in arr -> 인덱스의 값을 추출


arr=[4,2,3,5,1]
for i in range(0,5) :
    if arr[i]== 2 :
        print(f'{i}번째 요소가 일치')
        break
    else :
        i+=1
        if i==len(arr):
            print('찾을 수 없습니다.')
            break
            
#    elif i ==4 :
#       print('찾지 못했습니다.')
```

```python
# 함수사용

def linear_search(no,list) :
    for i in range(len(list)):
        if no==list[i]:
            return print(f'{i}번째 요소가 일치')
    return  print('찾지 못했습니다.')


print(linear_search(5,[4,2,3,5,1]))
```

```python
try:
    num = int(input('정수를 입력하세요 >> '))

except :
    print('정수를 입력하지 않았습니다.')
    
else:
    print('입력한 정수는 ',num,'입니다.')

finally :
    print('프로그램이 완전히 종료되었습니다.')
```

```python
def test():
  print('test()함수의 첫 줄입니다.')
  try:
    print('try  구문이 실행되었습니다.')
    return
    print('try 구문의 return 아래입니다.')
  except:
    print('except구문이 실행되었습니다.')
  else:
    print('else구문이 실행되었습니다.')
  finally:
    print('finally 구문이 실행되었습니다.')
  print('test()함수의 마지막 줄입니다.')

test()
```

```python
print('프로그램 실행')

while True:
    try :
        print('try 구문 실행')
        break
        print('try 구문의 break 아래')
    except:
        print('except 구문 실행')
    finally:
        print('finally 구문 실행')
    print('while의 마지막 줄')
print('프로그램 종료')
```
