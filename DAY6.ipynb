```python
# try except 
#if    

try:
    num_input = (input('  >'))
    
    num_a=int(num_input)
    
    print('  : ', num_a)
    print(' ', 2 * 3.14 * num_a)
    print(' ', 3.14 * num_a * num_a)

except: #  
    print(' ')
```
