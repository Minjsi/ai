```python
# 1. 알고리즘이란....
- 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표
  현한 아이디어 또는 생각
- 알고리즘 자체는 일상생활에서도 많이 이미 사용하고 있다. 
- 요리의 레시피, 음악의 악보, 가전제품 사용설명서 등
- 문자, 기호, 일러스트 등으로 나타낸다. 

2. 알고리즘과 프로그램
- 알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다. 
- 프로그래밍 언어(파이썬, 자바, 씨...)는 컴퓨터에 지시하기
  위한 인공 언어이다. 

3. 프로그램 작성에서의 알고리즘
- 프로그램 작성은 기획, 설계(*), 프로그래밍, 디버그 로 진행한다.
- 알고리즘은 설계 단계에서 필요하다. 

4. 좋은 알고리즘이란 ??
- 알기 쉽다. (쉬운 이해)
- 속도가 빠르다.
- 효율적이다.(메모리를 작게 사용한다.)
- 재이용(범용성)이 쉽다. (객체지향프로그래밍)

5. 왜 알고리즘을 공부해야 하나
- 좋은 프로그램을 만들기위해 (빠르고, 효율적, 범용성)
- 프로그램의 좋고 나쁨을 판단하기 위해
- 프로그램 작성 과정 전체를 효율화하기 위해
- 프로그램 기술을 향상시키기 위해 

6. 알고리즘의 조건
- 정확한 결과를 얻을 수 있어야 한다. 
  ( 알고리즘은 문제나 과제를 해결하기 위한 절차)
- 반드시 종료되어야 한다. 
  (중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다. )

7. 알고리즘의 기본형 3가지
- 순차 구조 : 처음부터 순서대로 처리하는 절차
  예) 심부름
      1 슈퍼마켓에 간다.
      2 라면을 산다.
      3 집으로 돌아온다. 
  실행을 원하는 처리를 위로부터 순서대로 작성한다. 
  알고리즘 중에서도 가장 많이 사용된다.
    
- 선택 구조 : 조건식으로 판단해 실행할 절차를 전환하는 구조

- 반복 구조 : 조건을 만족하는 동안 같은 처리를 반복하는 절차
```

```python
#반복문으로 팩토이얼 구하기

#함수선언
def facto(n):
    #변수선언
    output=1
    for i in range(1,n+1):
        output*=i
    return output

print('3! :', facto(3))
print('5! :', facto(5))
print('7! :', facto(7))
```

```python
# 재귀(recursion)로 팩토리얼 구하기
def facto(n):
  if n == 0:
    return 1 
  else:
    return n * facto(n-1)

print('3! :', facto(3))
print('3! :', facto(5))
print('3! :', facto(7))
#개발자들 사이에서는 재귀함수를 절대 사용하지 말자라는 의견도 있다.
# 그러나 가독성이 좋고 코드가 간결하고 멋스러워 보인다( 예쁜 쓰레기 )
```

```python
#피보나치수열(토끼의 번식 속도, 바이러스 감염 속도)
def fibo(n):
    if n==1 or n==2 :
        return 1
    else :
        return fibo(n-2)+fibo(n-1)
    
    
print('fibonacci(1) :' , fibo(1))
print('fibonacci(2) :' , fibo(2))
print('fibonacci(3) :' , fibo(3))
print('fibonacci(4) :' , fibo(4))
print('fibonacci(5) :' , fibo(5))
```
