```python
#선생님 정답

import random

def addition():
  num1 = random.randint(5,20)
  num2 = random.randint(5,20)
  print(f'{num1} + {num2} = ' )
  user_answer = int(input('Your answer : '))
  actual_answer = num1 + num2
  answers = (user_answer, actual_answer)
  return answers

def subtraction():
  num1 = random.randint(25,50)
  num2 = random.randint(1,25)
  print(f'{num1} - {num2} = ' )
  user_answer = int(input('Your answer : '))
  actual_answer = num1 - num2
  answers = (user_answer, actual_answer)
  return answers

def check_answer(user_answer, actual_answer):
  if user_answer == actual_answer:
    print('Correct')
  else:
    print('Incorrect, the answer is', actual_answer)

def main():
  print('1) Addition')
  print('2) Subtraction')
  selection = int(input('Enter 1 or 2 : '))
  if selection == 1:
    user_answer, actual_answer = addition()
    check_answer(user_answer, actual_answer)
  elif selection == 2:
    user_answer, actual_answer = subtraction()
    check_answer(user_answer, actual_answer)
  else:
    print('Incorrect selection')

main()  
```

```python

# 사용자가 이름 목록을 쉽게 관리할 수 있는 프로그램을 만듭니다.
# 목록에 이름을 추가할 수 있는 메뉴를 표시해야 합니다.
# 유저가 이름의 목록을 쉽게 관리 할수 있는 프로그램을 만든다. 
# 유저가 목록에 이름을 추가할 수 있는 메뉴와 목록의 이름을 수정하는 메뉴,
# 목록에서 이름을 삭제하는 메뉴, 목록의 모든 이름을 표시하는 메뉴를 화면에 표시한다. 
# 프로그램을 종료하는 메뉴도 있어야 한다. 사용자가 메뉴 외에 다른 것을 선택하면
# 적절한 메세지를 표시하자. 유저가 선택한 메뉴의 작업이 끝나면 다시 메뉴가 표시되도록 한다. 
# 가능한한 쉽게 프로그램을 만들자. 이. 프로그램을 만들어야 한다
# 가능한 한 사용하기 쉽습니다.
```

```python
def main():
    list=[]
    print('0)추가\n1)수정\n2)삭제\n3)모든이름 보기\n4)종료')
    n=input(input('select a num>>'))
    while True:
        if n==0:
            add_n(list)
            print(list)
        elif n==1:
            name=input('수정하고자 하는 이름을 입력하세요>>')
            rewrite(list,name)
            print(list)
        elif n==2:
            name=input('삭제하고자 하는 이름을 입력하세요')
            delete(list,name)
            print(list)
            


def add_n(list):
    t=True
    while t:
        name=input('추가할 이름 >> ')
        list.append(name)
        list.append
        a=input('더 추가? (y/n)>> ')
        if a==n:
            t=False
            return list
        
def rewrite(list,name):
    for i in list:
        if name==i:
            i==input('수정할 이름을 입력하세요>>')
            return list
        
def delete(list,name):
    for i in list:
        if name==i:
            del list[i]
            return list
        
main()
```
