```python
# pip install bs4
from bs4 import BeautifulSoup as bs #읽어온 웹페이지를 분석하기 쉽게 정제
```

```python
import pandas #데이터 분석 모듈
```

```python
html = requests.get("https://music.bugs.co.kr/chart") 
#웹페이지 코드 읽어온 것 코드를 그냥 글자로 가져온 것

#print(html.text)
```

```python
soup=bs(html.text) # beautifulSoup을 통해 데이터를 분석하기 용이하게 정제
print(soup)
```

```python
len(soup.select('tr')) #tr 태그가 붙은 부분들을 모두 추출
```

```python
len(soup.select('tbody > tr')) #원하는 100개가 아니므로 범위를 축소
```

```python
title= songs[0].select('p.title > a')[0].text #몇번째의 글자를 뽑아내라
title
```

```python
song_data=[]
rank=1
songs=soup.select('table.byChart > tbody > tr') #100곡 추출

for song in songs : 
    title=song.select('p.title>a')[0].text
    singer=song.select('p.artist>a')[0].text
    
    song_data.append([rank, title, singer])
    rank+=1
    
song_data
```

```python
import pandas as pd
df= pd.DataFrame(song_data, columns = ['순위','타이틀','가수'])
df
```

```python
from selenium import webdriver #파이썬으로 웹 브라우저 컨트롤
from bs4 import BeautifulSoup as bs #분석으로 용이하게 정제
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.genie.co.kr/chart/top200')
```

```python
html=driver.page_source
soup=bs(html)
soup
#songs=soup.select('table.byChart > tbody > tr')
```

```python
len(soup.select('tr.list'))
```

```python
songs=soup.select('tr.list')
song=songs[0]
song
```

```python
title= songs[0].select('a.title')[0].text.strip() #.strip() 잡동사니 날리기
title
```

```python
singer= song.select('a.artist')[0].text.strip()
singer
```

```python
song_data=[]
rank=1
songs=soup.select('tr.list')#50곡전체
for song in songs:
    title= songs[0].select('a.title')[0].text.strip()
    singer= song.select('a.artist')[0].text.strip()
    song_data.append([rank,title,singer])
    rank+=1
    
```

```python
import pandas as pd
df=pd.DataFrame(song_data,columns=['순위','타이틀','가수'])
df.to_excel('Genie.xlsx',index=False)
```
